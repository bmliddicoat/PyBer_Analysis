```python
# Set the x-axis to a list of strings for each month; Scatterplot.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create a scatterplot
plt.scatter(x_axis, y_axis, color='r', label='Chicago')

# Create labels for the x and y axes
plt.ylabel("(Fare($))")
plt.xlabel("Date")

# Create a title
plt.title("Pyber Fare by Month")

# Set limit for x-axis data
plt.ylim(0, 50)

#January to top
plt.scatter(x_axis, y_axis)
plt.gca().invert_yaxis()


# Add a legend
plt.legend()
```

```python
#Bubble Chart
plt.scatter(x_axis, y_axis)

#Change size of markers, use s, s parameter must be equal to feature of data
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjusting y_axis by any value
y_axis_larger= []
for data in y_axis:
    y_axis_larger.append(data*3)
    
#New y-axis list for s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refractor code: List comprehension new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Object-Oriented Interface for Scatter Plot

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Object-Oriented Bubble Chart
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Object-oreinted Appraoch, Bubble Chart
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent. alpha =
# Add a black edge color to the circles edgecolors= "black"
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, color="skyblue", s = [i * 5 for i in y_axis], alpha=.2, edgecolors= "black", label='Boston')

# name x & y axis
plt.xlabel("(Fare($))")
plt.ylabel("Date")

# Add a title.
plt.title("PyBer Fare by Month")


# Invert the y-axis so that January is at the top.
# Switch the axis so that the Fare($) data is on the x-axis.
ax.scatter(x_axis, y_axis)
plt.gca().invert_yaxis()

# Add a limit for the x-axis data.
plt.ylim(0, 50)
```

```python
# Pie Chart
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
#Add percentages for each month and "explode" the largest percentage, which is July/ figsize increases size of chart
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Add colors parameters assign a color variable to an array
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
